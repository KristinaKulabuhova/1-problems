```python
import pandas as pd
from graphviz import Digraph
```

Загрузим csv файл и посмотрим на него

```python
df = pd.read_csv('courses.csv')
df.head(50)
```

Удалим ненужные нам столбцы и строки

```python
df = df.drop(['Формат', 'Формат.1', 'Формат.2','Формат.3','Формат.4', 'Формат.5'], axis=1)
df = df.drop(list(range(21, 28)), axis=0)
```

```python
df.head(50)
```

Выгрузим список предметов

```python
def get_course_name(line):
    name = line[line.find(' ') + 1 :]
    deps_start = name.find('[')
    if deps_start == -1:
        return name
    else:
        return name[:deps_start - 1]

def get_course_no(line):
    return int(line[line.find('(') + 1: line.find(')')])

def get_course_deps(line):
    start = line.find('[')
    if start == -1:
        return None
    return line[start + 1 : line.find(']')].split(', ')
```

```python
courses = {}
for column in df.columns:
    for item in df[column]:
        if not pd.isna(item):
            course_no = get_course_no(item)
            courses[course_no] = {}
            courses[course_no]['course_name'] = get_course_name(item)
            courses[course_no]['course_deps'] = get_course_deps(item)
```

```python
courses
```

Теперь построим граф зависимостей предметов друг от друга

```python
graph = {key : [] for key in courses.keys()}

for key in courses.keys():
    deps = courses[key]['course_deps']
    
    if deps == None:
        continue
    
    for dep in deps:
        try:
            graph[key].append(int(dep))
        except ValueError:
            # В этом случае можно закрыть любой из предметов на выбор
            # Добавим фиктивную ноду, которая будет отражать этот факт
            # Так как в таблице нет предметов, которые бы имели зависимости вида [a, b, c/d, e/f]
            # то такая нода будет не более чем одна для каждого курса 
            # и мы можем для удобства дать ей номер -key
            graph[-key] = [int(x) for x in dep.split('/')]
            graph[key].append(-key)
```

```python
graph
```

Закодим графический дамп для графа зависимостей курсов. Для обозначения отношения "один из курсов" будем использовать фиктивные ноды (на графе обозначены как fictive for n)

```python
dot = Digraph(comment='Зависимости курсов')

for vert in graph.keys():
    if vert > 0:
        vert_name = '(' + str(vert) + ') ' + courses[vert]['course_name']
    else:
        vert_name = 'fictive for (' + str(-vert) + ')'
    dot.node(str(vert), vert_name)
    
for vert in graph.keys():
    for predecessor in graph[vert]:
        dot.edge(str(predecessor), str(vert))

dot.render('test-output/test-graph.png', view=True)  
    
```

Сделаем сначала простую реализацию: соединим курсы, которые нам необходимо пройти, ребрами с фиктивной вершиной потом пройдем рекурсивно вниз по всем необходимым курсам и так соберем список зависимостей

```python
def get_dependencies(g, course_no):
    deps = []
    
    deps.append(course_no)
    for course in g[course_no]:
        deps += get_dependencies(g, course)
    return deps

def clean_dep_list(dep_list):
    return [x for x in dep_list if x >= 0]

def get_course_names(dep_list):
    return list(set([courses[x]['course_name'] for x in dep_list]))

needed = [48, 50, 52, 59] # пример входных данных (необходимые курсы)
g = dict(graph)

g[-100] = needed

dependency_list = clean_dep_list(get_dependencies(g, -100))
subjects = get_course_names(dependency_list)
```

Начнем оптимизировать решение. В данный момент у нас не учитывается опциональность предметов, давайте начнем ее учитывать. Логично выбирать среди опциональных предметов те, которые имеют минимальное число зависимостей в целом и при этом максимальное число общих зависимостей с другими необходимыми нам предметами

Оптимизировать его можно следующим образом: сначала найдем ВСЕ предметы, которые надо закрыть, считая опциональные как обязательные. Затем создадим фиктивную ноду в начале нашего академического пути, к которой подвяжем все ноды нулевого уровня. Дальше у ребер, идущих из вершин без зависимостей сделаем capacity равным единице. У ребер, идущих из вершин с обязательными зависимостями сделаем capacity равным сумме capacity зависимостей. И наконец у ребер, идущих от вершин с опциональными зависимостями сделаем capacity равным максимальной сумме capacity зависимостей. Стоимостью прохода по ребру сделаем единицу, а потока пускать будем <количество предметов в семестр> * <количество семестров>

```python
class Edge:
    def __init__(self, fr, to, cap, cost):
        self.fr = fr
        self.to = to
        self.cap = cap
        self.cost = cost
    
    def __str__(self):
        return 'Edge {from: ' + str(self.fr) + ', to: ' + str(self.to) + ', cap: ' + str(self.cap) +', cost: ' + str(self.cost) + '}' 

def collect_edges(g, course_no, edges):
    if(len(g[course_no]) == 0):
        return 1
    
    sum_cap = 0
    for course in g[course_no]:
        cap = collect_edges(g, course, edges)
        sum_cap += cap
        
        edges.append(Edge(course, course_no, cap, 1))
    return sum_cap
        
edges = []
collect_edges(g, -100, edges)
print([str(x) for x in edges])
```
